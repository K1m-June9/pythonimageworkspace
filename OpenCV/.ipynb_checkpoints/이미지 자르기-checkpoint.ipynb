# 이미지 자르기

영역을 잘라서 새로운 윈도우에 표시

```python
import cv2
img = cv2.imread('img.jpg')
#img.shape #(391,640,3)

crop = img[100:200,300:400]#세로,가로

cv2.imshow('img',img)
cv2.imshow('crop',crop)

cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

영역을 잘라서 기존 윈도우에 표시

```python
import cv2
img = cv2.imread('img.jpg')

crop = img[100:200,200:400]#세로,가로
img[100:200, 400:600] = crop

cv2.imshow('img',img)

cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```
