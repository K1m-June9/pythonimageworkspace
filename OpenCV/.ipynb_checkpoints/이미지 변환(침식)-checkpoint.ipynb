# 이미지 변환(침식)

## 이미지를 깎아서 노이즈를 제거

```python
import cv2
import numpy as np

kernel = np.ones((3,3),dtype = np.uint8)

img = cv2.imread('reload.png',cv2.IMREAD_GRAYSCALE)

reload1 = cv2.erode(img,kernel,iterations=1)
reload2 = cv2.erode(img,kernel,iterations=2)
reload3 = cv2.erode(img,kernel,iterations=3)

cv2.imshow('1',reload1)
cv2.imshow('2',reload2)
cv2.imshow('3',reload3)

cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```
