# 이미지 회전

## 시계 방향 

```python
import cv2
img = cv2.imread('img.jpg')

rotate_90 = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE) #시계방향으로 90도 회전

cv2.imshow('90',rotate_90)

cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
import cv2
img = cv2.imread('img.jpg')

rotate_180 = cv2.rotate(img, cv2.ROTATE_180) #180회전

cv2.imshow('180',rotate_180)

cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```

## 반시계 방향

```python
import cv2
img = cv2.imread('img.jpg')

rotate_270 = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE) #반시계방향으로 90도 회전

cv2.imshow('90',rotate_270)

cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```
